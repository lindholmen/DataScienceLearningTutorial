```python
all_statistics = {}

def insert_dict(key):
    existing_value = all_statistics.setdefault(key, 0)
    all_statistics[key] = existing_value + 1

def decide_what_to_do(face):
    if face == 1:
        insert_dict(u'唱首歌')
    elif face == 2:
        insert_dict(u'跳个舞')
    elif face == 3:
        insert_dict(u'学狗叫')
    elif face == 4:
        insert_dict(u'转个圈')
    elif face == 5:
        insert_dict(u'念绕口令')
    else:
        insert_dict(u'讲冷笑话')

for i in range(10000):
    face = random.choice([1,2,3,4,5,6])
    decide_what_to_do(face)

result = pd.Series(all_statistics)
plt.title('plot bar chart based on Series')
plt.xlabel('Roll dice categories')
plt.ylabel('Freequency')
plt.grid(axis='y', alpha=0.75)
ax = result.plot(kind ="bar") # = result.plot.bar()

font = FontProperties(fname='/System/Library/Fonts/PingFang.ttc')
#font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=14) 
for label in ax.get_xticklabels() : 
    label.set_fontproperties(font) 
plt.show()
```

```python
df = pd.DataFrame({"山东省": [3872.18, 5960.42, 6650.02, 7162.2, 7662.1, 8542.44] , 
            "江苏省": [3057.39, 6104.21, 6480.34, 7099.95, 7997.82, 8182.727627]},
            index=["1995年", "1996年", "1997年", "1998年", "1999年", "2000年"])

df
```

```python
ax = df.plot(color=["g", "r"],style=["--","-"], title=u"GDP(单位:亿元)")
font = FontProperties(fname='/System/Library/Fonts/PingFang.ttc')

ax.set_xticks(range(len(df))) #never set the labels without setting the tick positions as well.
ax.set_xticklabels((label for label in df.index.values), fontproperties=font)

labels = ax.get_xticklabels()+ax.legend().texts+[ax.title]
for label in labels : 
    label.set_fontproperties(font) 
plt.show()
```
